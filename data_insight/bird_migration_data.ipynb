```python
def dropdown_options(col):
    return ['All'] + sorted(df[col].dropna().unique().tolist())

species_dropdown = widgets.Dropdown(
    options = sorted(df['Species'].unique()),
    description = 'Species:'
)
weather_dropdown = widgets.Dropdown(
    options = dropdown_options('Weather_Condition'),
    description = 'Weather:'
)
reason_dropdown = widgets.Dropdown(
    options = dropdown_options('Migration_Reason'),
    description = 'Reason:'
)
start_month_dropdown = widgets.Dropdown(
    options = dropdown_options('Migration_Start_Month'),
    description = 'StartMonth:'
)
end_month_dropdown = widgets.Dropdown(
    options = dropdown_options('Migration_End_Month'),
    description = 'EndMonth:'
)
food_level_dropdown = widgets.Dropdown(
    options = dropdown_options('Food_Supply_Level'),
    description = 'FoodSupply:'
)

output = widgets.Output()

def update_map(change = None):
    output.clear_output()

    filtered_df = df[df['Species'] == species_dropdown.value]

    if weather_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['Weather_Condition'] == weather_dropdown.value]
    if reason_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['Migration_Reason'] == reason_dropdown.value]
    if start_month_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['Migration_Start_Month'] == start_month_dropdown.value]
    if end_month_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['Migration_End_Month'] == end_month_dropdown.value]
    if food_level_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['Food_Supply_Level'] == food_level_dropdown.value]
    
    if not filtered_df.empty:
        map_center = [
            filtered_df[['Start_Latitude', 'End_Latitude']].values.mean(),
            filtered_df[['Start_Longitude', 'End_Longitude']].values.mean()
        ]
        m = folium.Map(location = map_center, zoom_start = 8)

        for _, row in filtered_df.iterrows():
            color = 'green' if row['Migration_Success'] == 'Successful' else 'red'
            points = [
                [row['Start_Latitude'], row['Start_Longitude']],
                [row['End_Latitude'], row['End_Longitude']]
            ]
            line = folium.PolyLine(locations = points, color = color, weight = 3, opacity = 0).add_to(m)
            PolyLineTextPath(
                line,
                text = '▶',
                repeat = True,
                offset = 8,
                attributes = {'fill': color, 'font-size': '10', 'stroke-width': '0.5'}
            ).add_to(m)
        
        with output:
            display(m)
    
    else:
        with output:
            print('선택한 종의 데이터가 없습니다.')

for widget in [species_dropdown, weather_dropdown, reason_dropdown,
               start_month_dropdown, end_month_dropdown, food_level_dropdown]:
    widget.observe(update_map, names = 'value')

filter_row1 = widgets.HBox([species_dropdown, weather_dropdown, reason_dropdown])
filter_row2 = widgets.HBox([start_month_dropdown, end_month_dropdown, food_level_dropdown])

display(filter_row1, filter_row2)
update_map()
display(output)
```
